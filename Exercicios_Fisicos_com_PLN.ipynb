```
import pandas as pd
import numpy as np

# Carregar o dataset
df = pd.read_csv("stretch_exercise_dataset new.csv")

# Substituir valores '0' nas colunas de interesse por NaN (valores ausentes)
columns_to_clean = ['Exercise Name', 'Preparation', 'Execution', 'Target_Muscles', 'Synergist_Muscles', 'Main_muscle']
df[columns_to_clean] = df[columns_to_clean].replace('0', np.nan)

# Verificar e remover linhas com dados ausentes em qualquer coluna
df_cleaned = df.dropna()

# Selecionar linhas
df_top_n = df_cleaned.head(47)
df_top_n.head(47)
```

```
# Função para preparar o texto para sumarização
def prepare_text_for_summarization(row):
    text_segments = []
    if pd.notna(row['Preparation']):
        text_segments.append(f"Preparation: {row['Preparation']}")
    if pd.notna(row['Execution']):
        text_segments.append(f"Execution: {row['Execution']}")
    return " ".join(text_segments)

# Aplicar a função para cada linha e gerar o texto para sumarização
texts_to_summarize = df_top_n.apply(prepare_text_for_summarization, axis=1).tolist()

texts_to_summarize[:47]
```

```
from langchain_community.chat_models import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.schema import Document

# Inicializar o modelo GPT-4 Mini
llm = ChatOpenAI(temperature=0, model_name="gpt-4o-mini")

# Definir o prompt para sumarização
prompt = ChatPromptTemplate.from_messages(
    [("system", "Write the most concise summary of the following:\n\n{context}")]
)

# Criar o chain de sumarização usando o prompt e o modelo GPT-4 mini
chain = create_stuff_documents_chain(llm, prompt)

# Função para preparar a entrada para o chain
def format_texts_for_chain(texts):
    # Converte cada texto em um objeto Document
    docs = [Document(page_content=text) for text in texts]
    return docs

# Função para realizar a sumarização usando LangChain
def summarize_text_with_langchain(texts):
    docs = format_texts_for_chain(texts)
    result = chain.invoke({"context": docs})
    return result

# Aplicar a sumarização nas linhas e exibir os resumos para cada uma
for idx, text in enumerate(texts_to_summarize):
    summary = summarize_text_with_langchain([text])
    print(f"Summary for line {idx + 1}: {summary}")
```

O método de extração de palavras-chaves envolve a aplicação de um modelo de linguagem para identificar termos relevantes em um texto a partir de um prompt específico. No entanto, não requer um processamento manual do texto, mas utiliza um modelo treinado para detectar automaticamente as palavras mais significativas.
Assim, vamos utilizar a classe ***PromptTemplate*** para criar o prompt e, em seguida, utilizamos um prompt ***template*** que visa instruir a extração das palavras-chave e orienta o modelo a focar na identificação dos termos centrais do texto

## Sistema de sugestões de exercícios

Essa técnica utiliza uma abordagem interativa e baseada em pesquisa para recomendar exercícios específicos com base no músculo escolhido pelo usuário de seu interesse a partir da capturação e do armazenamento do input ***user_input***, na qual filtra um conjunto de dados previamente carregado e exibindo as informações relevantes.

```
# Solicitar ao usuário que insira um músculo
user_input = input("Type a muscle to search: ")
```

Em seguida, o input é usado para filtrar um DataFrame ***df_top_n***, onde contem informações sobre diferentes exercícios em que, tal filtragem, fica verificando se o nome do músculo escolhido pelo usuário está presente em qualquer uma das colunas *Target_Muscles*, *Synergist_Muscles* ou *Main_muscle*. E, finalmente, o sistema cria uma lista ***result*** para armazenar os exerícios que correspondem ao músculo selecionado e exibe os resultados formatados.

```
# Filtrar o DataFrame com base na palavra-chave inserida pelo usuário
filtered_df = df_top_n[df_top_n['Target_Muscles'].str.contains(user_input, case=False, na=False) |
                         df_top_n['Synergist_Muscles'].str.contains(user_input, case=False, na=False) |
                         df_top_n['Main_muscle'].str.contains(user_input, case=False, na=False)]

# Criar uma lista para armazenar os resultados formatados
results = []

# Formatar a saída para a lista
if not filtered_df.empty:
  for index, row in filtered_df.iterrows():
    exercise_name = row['Exercise Name']
    preparation = row['Preparation']
    execution = row['Execution']
    results.append({
        "Exercise name": exercise_name,
        "Preparation": preparation,
        "Execution": execution
    })

  # Exibir os resultados formatados
  print(f"\nExercises for the selected muscle '{user_input}':\n")
  for result in results:
    print(result)
else:
  print(f"\nNo exercise found for muscle '{user_input}'.")
```
